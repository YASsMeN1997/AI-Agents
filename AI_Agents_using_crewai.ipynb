````python
import os
from crewai import Agent, Task, Crew, Process,LLM
from crewai.knowledge.source.string_knowledge_source import StringKnowledgeSource
from langchain_openai import ChatOpenAI
from pydantic import BaseModel
from google.colab import userdata
````

````python

os.environ["GOOGLE_API_KEY"] = userdata.get('GOOGLE_API_KEY')

llm = LLM(
              model='gemini/gemini-1.5-flash',
              api_key=os.environ["GOOGLE_API_KEY"]
            )
````

````python
import agentops
from google.colab import userdata

os.environ['AGENTOPS_API_KEY'] = userdata.get('agent_test')
agentops.init(
    api_key=userdata.get('agent_test'),
    skip_auto_end_session=True
)
````

````python
output_dir = './ai-agent-output'
os.makedirs(output_dir, exist_ok=True)
````

````python
about_company = "YZ company is a company that provides AI soluations to help websites refine thier search and recommendation systems."
company_context = StringKnowledgeSource(content = about_company)
````

## setup agent
- agent a: to provide a list of suggested search queries to be passed to the search engine.

````python
from pydantic import BaseModel, Field
from typing import List, Optional

num_of_keywors = 100

class SuggestedSearchQueries(BaseModel):
  queries: List[str] = Field(...,title="A list of suggested search queries to be passed to the search engine.",
                             min_items=1,max_items=num_of_keywors)

search_queries_recomendation_agent = Agent(
    role = "Search Queries Recomendation Agent",
    goal = "\n".join([
        'To Provide a list of suggested search queries to be passed to the search engine.',
        'the queries must be veried and looking for specific items'
    ]

    ),
    backstory='the agent is designed to help in looking for products by providing a list of suggested search queries to be passed to the search engine based on the context provided',
    llm=llm,
    verbose=True

)

search_queries_recomendation_task = Task(
    description="\n".join([
        'YZ company is looking to buy {product_name} at the best prices(value for a price strategy)',
        'the company target must be any of the websites to buy from: {websites_list}',
        'the company wants to reach all avilable products on the internet to be compared later in another stage.',
        'the stores must sell the product in {country_name}',
        'generate at maximum {num_of_keywors} queries.',
        'the search keywords must be in {language} language.',
        'the search query must reach an ecommerce webpage for product, and not a blog or listing page.'
    ]

    ),
    expected_output='A JSON object containing a list of suggested search queries.',
    output_json=SuggestedSearchQueries,
    output_file=os.path.join(output_dir,'suggested_search_queries.json'),
    agent=search_queries_recomendation_agent,

)
````

- run the crew

````python
yz_crew = Crew(agents=[search_queries_recomendation_agent],
               tasks=[search_queries_recomendation_task],
               process=Process.sequential)
````

````python
crew_res = yz_crew.kickoff(
    inputs={
        "product_name": "coffee machine for the office",
        "websites_list": ['www.amazon.com','www.jumia.com','www.noon.com'],
        "country_name": "EGYPT",
        "language": "arabic",
        "num_of_keywors": 10,
    }
)
````

````python
from tavily import TavilyClient

search_client = TavilyClient(api_key=userdata.get('tavily_api_key'))
````

````python
search_client.search('coffee machine site: amazon.eg')
````

## agent b
- to search (GOOGLE) for products based on the suggested search query.

````python
from crewai.tools import tool
````

````python
class singlesearchresult(BaseModel):
  title: str
  link: str
  content: str
  confidence_score: float
  search_query: str

class AllSearchResults(BaseModel):
  results: List[singlesearchresult]
````

````python
@tool
def search_engine_tool(query:str):
  """useful for search -based queries. use this tool to find current information about any query related pages using a search engine"""
  return search_client.search(query)

search_engine_agent = Agent(
     role = "Search engine agent",
    goal = "search for products based on the suggested search query",
    backstory='the agent is designed to help in looking for products by providing a list of suggested search queries .',
    llm=llm,
    verbose=True,
     tools=[search_engine_tool]
)

search_engine_task = Task(
        description="\n".join([
        'the search is to search for products based on the suggested search queries.',
         'you have to collect results from multiple search queries.',
        'ignore any susbicious links or not an ecommerce single product website link.',
        'ignore any search results with confidence score less than ({score_th})',
        'the search results will be used to cimpare prices of products from different websites.'
    ]

    ),
    expected_output='A JSON object containing the search results.',
    output_json=AllSearchResults,
    output_file=os.path.join(output_dir,'step1_search_results.json'),
    agent=search_engine_agent,
)
````

````python
yz_crew = Crew(agents=[search_queries_recomendation_agent,
                       search_engine_agent],
               tasks=[search_queries_recomendation_task,
                      search_engine_task],
               process=Process.sequential)
````

````python
crew_res = yz_crew.kickoff(
    inputs={
        "product_name": "coffee machine for the office",
        "websites_list": ['www.amazon.com','www.jumia.com','www.noon.com'],
        "country_name": "EGYPT",
        "language": "arabic",
        "num_of_keywors": 10,
        'score_th': 0.60,

    }
)
````

## agent c
- to extract product details from any web page.

````python
from scrapegraph_py import Client
import json
scrap_client = Client(api_key = userdata.get('scrapgraph'))
````

````python
class ProductSpecs(BaseModel):
  name: str
  value: str

class Single_extracted_product(BaseModel):
  page_url: str = Field(...,title="The url of the page where the product was found.")
  product_title: str = Field(...,title="The title of the product.")
  product_price: str = Field(...,title="The price of the product.")
  product_image_url: str = Field(...,title="The url of the product image.")
  product_url: str = Field(...,title="The url of the product.")
  product_current_price: float = Field(...,title="The current price of the product.")
  product_original_price: float = Field(title="The original price of the product before discount. set to none if no discount",default=None)
  product_discount_percentage: float = Field(title="The discount percentage of the product.set to none if no discount",default=None)
  product_specs: List[ProductSpecs] = Field(...,title="A list of product specifications. focus on the most important specs",min_items=1,max_items = 5)
  agent_recommendation_rank: int = Field(...,title="The rank of the product to be considered in the final procurement report. (out of 5, higher is better) in the recommendation list ordering from best to the worst.")
  agent_recommendation_notes: List[str] = Field(...,title="A set of notes why would you recommend or not recommed this product to the compny, compared to other products.")

class AllExtractedData(BaseModel):
  products:List[Single_extracted_product]
````

````python
@tool
def WebScrapTool(page_url:str):
  """
  An Ai tool to help an agent to scrape a web page

  example:
  WebScrapTool(
    page_url:"https://www.amazon.eg/nespresso-machine/s?k=nespresso+machine",
  )
  """
  details = scrap_client.smartscraper(
      website_url = page_url,
      user_prompt= 'extract ```json\n' + Single_extracted_product.schema_json() +"```\n from the web page")
  return {
      "page_url": page_url,
      "details": details
  }

scrapping_agent = Agent(
    role = "web scrapping agent",
    goal = "to extract details from any websites",
    backstory='the agent is designed to help in looking for required values from any website url. this details will be used to decide which best product to buy.',
    llm=llm,
    verbose=True,
    tools=[WebScrapTool]
)

scrapping_task = Task(
    description="\n".join([
        'the task is to extract product detals from any ecommerece store page url.',
        'the task has to collect results from multiple pages urls.',
        'collect the best {top_recommendations_no} products from the search results.'

    ]),
    expected_output='A JSON object containing the products details.',
    output_json=AllExtractedData,
    output_file=os.path.join(output_dir,'step3_products_results.json'),
    agent=scrapping_agent,
)
````

````python
yz_crew = Crew(agents=[search_queries_recomendation_agent,
                       search_engine_agent,
                       scrapping_agent],

               tasks=[search_queries_recomendation_task,
                      search_engine_task, scrapping_task],
               process=Process.sequential)

crew_res = yz_crew.kickoff(
    inputs={
        "product_name": "coffee machine for the office",
        "websites_list": ['www.amazon.com','www.jumia.com','www.noon.com'],
        "country_name": "EGYPT",
        "language": "arabic",
        "num_of_keywors": 10,
        'score_th': 0.10,
        "top_recommendations_no": 10

    }
)
````

## agent d
- to generate a professional procurement report.

````python
procurement_report_agent = Agent(
    role = "Procurement report author agent",
    goal = "to generate a professional HTML page for the procurement report",
    backstory='the agent is designed to assist in generating a professional HTML page for the procurement report after looking into a list of products.',
    llm=llm,
    verbose=True
    # no need to tool here.
)

procurement_report_task = Task(
    description="\n".join([
        'the task is to generate a professional HTML page for the procurement report.',
        'you have to use Bootstrap CSS framework for a better UI.',
        'use the provided context about the company to make a specilized report.',
        'the report will include the search results and prices of products from different websites.',
        'the report will include the following sections:',
        '1. **Executive Summary:** Briefly summarize the procurement process and key findings.',
        '2. **Product Requirements:** Outline the specifications and criteria for the desired product (e.g., coffee machine for the office).',
        '3. **Market Analysis:** Present an overview of the market research conducted, including websites considered (e.g., Amazon, Jumia, Noon) and search queries used.',
        '4. **Product Comparison:** Provide a detailed comparison of the shortlisted products, including features, prices, and vendor information.',
        '5. **Recommendations:** Offer clear recommendations on the preferred product(s) based on the evaluation criteria.',
        '6. **Appendix:** Include any supporting documentation, such as product datasheets or vendor quotes.'

    ]),
    expected_output='A Professional HTML page for the procurement report.',
    output_file=os.path.join(output_dir,'step_4_procurement_report.html'),
    agent=procurement_report_agent,
)
````

## Run All Crew

````python
yz_crew = Crew(agents=[search_queries_recomendation_agent,
                       search_engine_agent,
                       scrapping_agent,
                       procurement_report_agent],

               tasks=[search_queries_recomendation_task,
                      search_engine_task,
                      scrapping_task,
                      procurement_report_task],

               process=Process.sequential,
               knowledge_source = company_context)

````

````python
crew_res = yz_crew.kickoff(
    inputs={
        "product_name": "coffee machine for the office",
        "websites_list": ['www.amazon.com','www.jumia.com','www.noon.com'],
        "country_name": "dubai",
        "language": "english",
        "num_of_keywors": 10,
        'score_th': 0.10,
        "top_recommendations_no": 10

    }
)
````
